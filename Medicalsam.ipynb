```python
from google.colab import drive
import os
import requests
import shutil

# Step 1: Mount Google Drive
drive.mount('/content/drive')

# Step 2: Define the URLs and paths
zip_urls = {
    'test_data': 'https://challenge.isic-archive.com/data/ISBI2016_ISIC_Part1_Test_Data.zip',
    'training_data': 'https://challenge.isic-archive.com/data/ISBI2016_ISIC_Part1_Training_Data.zip',
    'test_groundtruth': 'https://challenge.isic-archive.com/data/ISBI2016_ISIC_Part1_Test_GroundTruth.csv',
    'training_groundtruth': 'https://challenge.isic-archive.com/data/ISBI2016_ISIC_Part1_Training_GroundTruth.csv'
}

destination_dir = '/content/drive/MyDrive/ISIC/data/isic'

# Ensure the destination directory exists
os.makedirs(destination_dir, exist_ok=True)

# Step 3: Download and unzip the files
for file_type, url in zip_urls.items():
    local_path = os.path.join('/content', os.path.basename(url))

    # Download the file
    if not os.path.exists(local_path):
        print(f'Downloading {file_type}...')
        r = requests.get(url, allow_redirects=True)
        open(local_path, 'wb').write(r.content)

    # Unzip the file if it's a zip archive
    if local_path.endswith('.zip'):
        print(f'Unzipping {file_type}...')
        !unzip -q "$local_path" -d "$destination_dir"
    else:
        # Move the CSV files
        print(f'Moving {file_type}...')
        shutil.move(local_path, os.path.join(destination_dir, os.path.basename(url)))

# Step 4: Verify the contents of the destination directory
print("Contents of the destination directory:", os.listdir(destination_dir))
```
